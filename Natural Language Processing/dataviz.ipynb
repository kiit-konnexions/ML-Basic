```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv', encoding="latin-1")
df.head()
```

```python
df.dtypes
```

```python
# display features having missing values
[col for col in df.columns if df[col].isnull().sum() > 0]
```

```python
df_country = pd.read_excel('Country-Code.xlsx')

# Join the two dataframes
df = pd.merge(df, df_country, on='Country Code', how='left')
df.dtypes
```

```python
# Plot pie chart on country name vs count
df['Country'].value_counts().plot.pie()
```

```python
# Plot for only top 4 countries with % of total
ex = (0, 0.1, 0, 0)
df['Country'].value_counts().head(4).plot.pie(explode=ex, autopct='%1.1f%%')
```

```python
sns.barplot(x='Aggregate rating', y='Rating text', data=df)
```

```python
# Which currency is used in which country and it's plot country vs currency

df.groupby(['Country', 'Currency']).size().unstack().plot(kind='bar', stacked=True)
```

```python
# List countries which have online delivery
df[df['Has Online delivery'] == 'Yes']['Country'].unique()
```

```python
# List countries which have zero rating
df[df['Aggregate rating'] == 0]['Country'].unique()
```

```python
# Combine aggregate rating and rating color and rating text
df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index(name='Count')
```
